```python
import os

import time
import torch
import datetime
import brian2 as b2
from brian2 import *
from itertools import chain
from functools import partial
import matplotlib.pyplot as plt

from torch.utils.data import Dataset, DataLoader
import numpy as np

from dataset_util import Synthetic_Dataset_Utils




from brian2 import prefs
#from brian2 import set_device
from brian2.devices import device_module

import brian2cuda
from brian2cuda import *

import brian2tools as b2t

from brian2 import prefs


prefs.codegen.target = 'cython'



#set_device("cpp_standalone", directory='project', build_on_run=False)

#device = b2.get_device()


#set_device("cuda_standalone", build_on_run=True)
# Set specific preferences for CUDA standalone


prefs.devices.cuda_standalone.cuda_backend.cuda_runtime_version = (12.2)
prefs.devices.cuda_standalone.profile_statemonitor_copy_to_host = None
prefs.devices.cuda_standalone.cuda_backend.extra_compile_args_nvcc = ['-use_fast_math', '-w']
prefs.devices.cuda_standalone.cuda_backend.detect_gpus = True
prefs.devices.cuda_standalone.cuda_backend.gpu_id = 0  # Use the first GPU
# prefs.devices.cuda_standalone.cuda_backend.cuda_path = '/home/brovoletto/SerotoninSNN/nvcc_installation/nvcc_folder'



# REMEMBER YOU SWITCHED SOME EXCITATORY WITH INHIBITORY!!!!!!!!


###############################################################################
# Model Parameters and Equations
###############################################################################
msec_step = 400 * ms

# TODO: Make weights always non-negative!


# Equations for everything. I_post in the synapse is modifying I in the target neuron


# TODO: Make weights always non-negative! (maybe not necessary)
# Equations for everything. I_post in the synapse is modifying I in the target neuron



# Parameters for neurons
Cm = 281 * pF  # Membrane capacitance
gL = 30 * nS   # Leak conductance
EL = -70.6 * mV  # Leak reversal potential
Vth = -45.4 * mV  # Spike threshold
Vth_out = -25.4 * mV  # Spike threshold

DeltaT = 5 * mV  # Slope factor
Vr = -70.6 * mV  # Reset potential

Vcut = -40 * mV  # Cutoff potential for spike generation
tau_A = 300 * ms  # Adaptation time constant
c = 10 * nS       # Coupling parameter
b = 2 * nA  # Spike-triggered adaptation increment

# Parameters for synapses
tau_rise = 5 * ms  # Rise time constant for AMPA
tau_decay = 50 * ms  # Decay time constant for NMDA
sigma_w_init = 3 * nsiemens   # Initial synaptic weight (conductance)
w_init_2 = 3 * nsiemens

regularizer = 10 * nvolt



# Parameters for neurons
Cm = 281 * pF  # Membrane capacitance
gL = 30 * nS   # Leak conductance
EL = -70.6 * mV  # Leak reversal potential
Vth = -45.4 * mV  # Spike threshold
DeltaT = 2 * mV  # Slope factor
Vr = -70.6 * mV  # Reset potential
Vcut = -40 * mV  # Cutoff potential for spike generation
tau_A = 1 * ms  # Adaptation time constant
c = 4 * nS       # Coupling parameter
b = 1 * nA  # Spike-triggered adaptation increment

# Parameters for synapses
tau_rise = 5 * ms  # Rise time constant for AMPA
tau_decay = 50 * ms  # Decay time constant for NMDA
w_init = 0.5 * nS  # Initial synaptic weight (conductance)

# LSD-modulated parameters
DeltaT = 3 * mV
gL= 20 * nS
Vth = -52 * mV
b = 1 * nA
tau_A = 14.05 * ms


# AdEx neuron equations. Separate terms are used for each I contribution, in order to work in brian2. Inhibitory currents are inhibitory
# simply by means of the minus sign.

# Spike trace terms (X and Y) are implemented following the equetions described in "Predictive coding with spiking neurons and feedforward gist signaling".
eqs = '''
dv/dt = ( -gL * (v - EL) + gL * DeltaT * exp((v - Vth) / DeltaT) + I_tot - A ) / Cm : volt
dA/dt = (c * (v - EL) - A) / tau_A : amp
dX/dt = -X / tau_decay + Y / tau_rise : volt    # Spike trace
dY/dt = -Y / tau_decay : volt  # Glutamate decay
Iexc : amp
Iexc_2 : amp
Iexc_3 : amp
Iexc_4 : amp
Iinh : amp
Iinh_2 : amp
Iinh_3 : amp
Iinh_4 : amp
I_tot = Iexc + Iexc_2 + Iexc_3 + Iexc_4 - Iinh - Iinh_2 - Iinh_3 - Iinh_4 : amp
batch_sum_X : volt   # Accumulate X values over a batch

'''
# Ii = I + input_stimuli(t)  :  amp  # Input current

# These take inputs from the input_stimuli TimedArray
input_eqs = '''
dv/dt = ( -gL * (v - EL) + gL * DeltaT * exp((v - Vth) / DeltaT) + I - A ) / Cm : volt
dA/dt = (c * (v - EL) - A) / tau_A : amp
dX/dt = -X / tau_decay + Y / tau_rise : volt    # Spike trace
dY/dt = -Y / tau_decay : volt  # Glutamate decay
I = input_stimuli(t, i) : amp
indices : integer
'''

# These take inputs from output_stimuli TimedArray
out_eqs = '''
dv/dt = ( -gL * (v - EL) + gL * DeltaT * exp((v - Vth) / DeltaT) + I_tot - A ) / Cm : volt
dA/dt = (c * (v - EL) - A) / tau_A : amp
dX/dt = -X / tau_decay + Y / tau_rise : volt    # Spike trace
dY/dt = -Y / tau_decay : volt  # Glutamate decay
Iexc_internal : amp  # Synaptic input
Iexc_3 : amp
Iexc_4 : amp
Iinh_3 : amp
Iinh_4 : amp
I_tot = Iexc_3 + Iexc_4 - Iinh_3 - Iinh_4 + Iexc_internal + output_stimuli(t, i) : amp
indices : integer  # dimensionless index variable
spike_count = 0 : integer
batch_sum_X : volt
'''

# Synapse equations (Spike trace dynamics)

# Each synapse has a weight parameter, which determines how much its spike trace X influences the current I of the postsynaptic neuron.
syn_eqs_exc = '''
w : siemens   # Synaptic weight, conductance

Iexc_post = w * X_pre : amp (summed)
'''

syn_eqs_exc_2 = '''
w : siemens   # Synaptic weight, conductance

Iexc_2_post = w * X_pre : amp (summed)
'''

syn_eqs_exc_3 = '''
w : siemens

Iexc_3_post = w * X_pre : amp (summed)
'''

syn_eqs_exc_4 = '''
w : siemens

Iexc_4_post = w * X_pre : amp (summed)
'''


syn_eqs_exc_output = '''
w : siemens   # Synaptic weight, conductance

Iexc_internal_post = w * X_pre  : amp (summed)
'''



syn_eqs_inh= '''
w : siemens   # Synaptic weight, conductance

Iinh_post = w * X_pre  : amp (summed)
'''

syn_eqs_inh_2= '''
w : siemens   # Synaptic weight, conductance

Iinh_2_post = w * X_pre  : amp (summed)
'''

syn_eqs_inh_3= '''
w : siemens   # Synaptic weight, conductance

Iinh_3_post = w * X_pre : amp (summed)
'''

syn_eqs_inh_4= '''
w : siemens   # Synaptic weight, conductance

Iinh_4_post = w * X_pre  : amp (summed)
'''


# As in the paper, when a neuron spikes it causes an immediate change in the value of Y and A.

on_pre_value = 'Y_post = 230*mvolt; A_post = A_post + b'


# This function takes a dims list and makes len(dims) neuron groups, of the sizes specified in dims
def make_groups(dims, eqs=eqs, with_input=False):
    groups = []
    for ii, dim in enumerate(dims):
        if ii == 0 and with_input:
            g = NeuronGroup(dim, input_eqs, threshold='v > -40*mV', reset='v = Vr', method='euler')
        else:
            g = NeuronGroup(dim, eqs, threshold='v > -40*mV', reset='v = Vr', method='euler')
        g.v = EL
        groups.append(g)
    return groups


# We add a custom_name attribute to synapses for debugging purposes, in order to identify them more easily since brian2 changes their name.
class NamedSynapses(Synapses):
    def __init__(self, source, target=None, model=None, custom_name=None, **kwargs):
        super().__init__(source, target, model, **kwargs)
        
        self.add_attribute("brian2_name")
        self.add_attribute("custom_name")
        
        self.brian2_name = self.name  
        
        self.custom_name = custom_name or f"{source}-{target}"
        
        print(self.brian2_name, self.custom_name)
        
    def connect(self, **kwargs):
        # Call the original connect method
        super().connect(**kwargs)
        if self.custom_name == "suygu5":
            # Print the connections
            connections = [(ii, j) for ii, j in zip(self.i[:], self.j[:])]
            print("Connections created:")
            for pre, post in connections:
                print(f"Pre: {pre}, Post: {post}")
            
        
# Same for the monitors
class NamedSpikeMonitor(SpikeMonitor):
    def __init__(self, group, custom_name=None, **kwargs):
        super().__init__(group, **kwargs)
        self.add_attribute("custom_name")
        self.custom_name = custom_name or str(group)    
     
     
# This function can be useful when using brian2cuda since .connect() may not work correctly. Not used here.
def manual_fully_connect(syn, N_pre, N_post):
    # Manually specify connections
    pre_indices = [ii for ii in range(N_pre) for _ in range(N_post)]
    post_indices = [j for _ in range(N_pre) for j in range(N_post)]

    # Connect manually
    syn.connect(i=pre_indices, j=post_indices) 
    
    return syn  
            

# This function is called to make bottom-up connections. They are one-to-one (not fully connected) and they do not evolve over time.
# We are following "Predictive coding with spiking neurons and feedforward gist signaling".
# They connect the representation unit Rs[n] with the error units Es_0[n] (excitatory) and Es_1[n] (inhibitory) of the same area
def make_bottom_up_connections(R, E_0, E_1):
    print("A")
    print("Bottom-up: ", R.N, E_0.N, E_1.N )
    S_p = NamedSynapses(R, E_0, model=syn_eqs_exc, on_pre= on_pre_value, custom_name = "s1")
    S_p.connect(j='i')
    S_p.w = 'w_init_2'

    S_m = NamedSynapses(R, E_1, model=syn_eqs_inh, on_pre= on_pre_value, custom_name = "s2")
    S_m.connect(j='i')
    S_m.w = 'w_init_2'

    return S_p, S_m



# Name is slightly misleading. They are one to one connections paired with the ones from the function above, but signal in the opposite direction.
def make_bottom_up_connections_2(E_0, E_1,R):
    print("B")
    print("Bottom-up-2: ", R.N, E_0.N, E_1.N )
    S_p = NamedSynapses(E_0, R, model=syn_eqs_inh_2, on_pre= on_pre_value, custom_name = "s3")
    S_p.connect(j='i')
    S_p.w = 'w_init_2'

    S_m = NamedSynapses(E_1, R,  model=syn_eqs_exc_2, on_pre= on_pre_value, custom_name = "s4")
    S_m.connect(j='i')
    S_m.w = 'w_init_2'

    return S_p, S_m


# This function makes fully connected synapses which then evolve during training. 
# They connect the error units Es_0[n] (excitatory) and Es_1[n] (inhibitory) with the representation unit Rs[n+1] of the area above.
# abs(randn()) creates a half gaussian with only positive values.

def make_top_down_connections(R, E_0, E_1):
    print("C")
    print("Top-down: ", R.N, E_0.N, E_1.N )
    
    S_p = NamedSynapses(R, E_0, model=syn_eqs_inh_3, on_pre= on_pre_value, custom_name = "s6")
    S_p.connect()
    #manual_fully_connect(S_m, R.N, E_0.N)
    S_p.w = 'abs(randn()) * sigma_w_init'
    
    
    S_m = NamedSynapses(R, E_1, model=syn_eqs_exc_3, on_pre= on_pre_value, custom_name = "s5")
    S_m.connect()
    #manual_fully_connect(S_p, R.N, E_1.N)
    S_m.w = 'abs(randn()) * sigma_w_init'



    return S_p, S_m


# As the function above, but in the opposite direction.
def make_top_down_connections_2(E_0, E_1, R):
    print("D")
    print("Top-down-2: ", R.N, E_0.N, E_1.N )
    
    S_p_2 = NamedSynapses(E_0, R,  model=syn_eqs_exc_4, on_pre= on_pre_value, custom_name = "s8")
    S_p_2.connect()
    #manual_fully_connect(S_m_2, E_0.N, R.N)
    S_p_2.w = 'abs(randn()) * sigma_w_init'
    
    
    S_m_2 = NamedSynapses(E_1, R,  model=syn_eqs_inh_4, on_pre= on_pre_value, custom_name = "s7")
    S_m_2.connect()
    #manual_fully_connect(S_p_2, E_1.N, R.N)
    S_m_2.w = 'abs(randn()) * sigma_w_init'



    return S_p_2, S_m_2


# Gist connections as in the paper. They connect Rs[0] with every other representation unit. Weights are fixed, not learned.
def make_gist_connections(Rs, G):
    S_gist_input = NamedSynapses(Rs[0], G, model=syn_eqs_exc, on_pre= on_pre_value, custom_name = "s9")
    S_gist_input.connect(p=0.05)
    S_gist_input.w = 'abs(randn()) * sigma_w_init * 2'

    # Connect G to all other layers
    S_gist_output = NamedSynapses(G, Rs[-1], model=syn_eqs_exc, on_pre= on_pre_value, custom_name = "s10")
    S_gist_output.connect(p=0.05)
    S_gist_output.w = 'abs(randn()) * sigma_w_init * 2'
    return S_gist_input, S_gist_output


# The output layer mirrors the possible hierarchical classes in the dataset. Therefore, if the dataset-creation parameters num_classes_per_layer = 3 and max_depth = 4 
# Then, the output layer is going to mirror the dataset's classes, by having 4 layers of size [3, 3^2, 3^3, 3^4]. The reason is that each child class
# produces a number of child classes equal to num_classes_per_layer.

def make_output_layer(num_classes_per_layer, max_depth):
    
    '''
    if w_out_init is None:
        w_out_init = w_init/3000
        print("W_OUT_INIT: ", w_out_init)
    '''

    total_neurons = sum([num_classes_per_layer**(d) for d in range(max_depth)])
    Os = NeuronGroup(N=total_neurons, model=out_eqs, threshold='v > -40*mV', reset='v = Vr', method='euler')

    #S_o = NamedSynapses(Os, Os, model=syn_eqs_exc_output, on_pre= on_pre_value)

    trace = 0
    for d in range(num_classes_per_layer - 1):
        new_trace = trace + num_classes_per_layer**(d+1)
        source_indices = list(range(trace, trace + num_classes_per_layer**(d+1)))
        target_indices = list(range(new_trace, new_trace + num_classes_per_layer**(d+2)))
        ii, jj = np.meshgrid(source_indices, target_indices, indexing='ij')
        #S_o.connect(i=ii.flatten(), j=jj.flatten())
        trace = new_trace

    # Assign indices for output stimuli
    stimulus_indices = []
    for ii in range(max_depth):
        for _ in range(num_classes_per_layer**(ii+1)):
            stimulus_indices.append(ii)
    Os.indices = stimulus_indices

    return Os#, S_o


# This function uses the previously defined functions to construct the whole network. len(dims) determines the depth of the network, and the
# values inside dims determine the size of each layer. Each value corresponds to the size of Rs[n] and Es_0[n] and Es_1[n].

def make_network(dims, num_classes_per_layer, max_depth):
    max_depth += 1

    # Using dims, we make representation layers and corresponding excitatory (0) and inhibitory (1) Error layers.
    Rs = make_groups(dims, with_input=True)
    Es_0 = make_groups(dims)
    Es_1 = make_groups(dims)
    # Gist neurons, 16 as in the paper
    G = NeuronGroup(16, eqs, threshold='v > -40*mV', reset='v = Vr', method='euler')

    # Optional synapses inside the outpt layer, to connect parent neurons to child neurons.
    #Os, S_o_internal = make_output_layer(num_classes_per_layer, max_depth, w_out_init=w_out_init)
    Os = make_output_layer(num_classes_per_layer, max_depth)
    #w_out_init = w_init/100
    
    connections = {}
    for l in range(len(Rs)):
        
        S_p, S_m = make_bottom_up_connections(Rs[l], Es_0[l], Es_1[l])
        connections[f"bottom_up_{l}"] = [S_p, S_m]

        # Backward one-to-one connections are NOT created from Rs[1] to Rs[0] (the input layer)
        if l != 0:
            
            
            S_p_2, S_m_2 = make_bottom_up_connections_2(Es_0[l], Es_1[l], Rs[l])
            connections[f"bottom_up_2_{l}"] = [S_p_2, S_m_2]
        
        # top down connections between representation unit and error units of the previous layer. 
        # if l = len(Rs), the error units need to connect to the output layer (Os). This happens below.
        if l != len(Rs)-1:

            S_p_td, S_m_td = make_top_down_connections(Rs[l+1], Es_0[l], Es_1[l])
            S_p_td_2, S_m_td_2 = make_top_down_connections_2(Es_0[l], Es_1[l], Rs[l+1])
            connections[f"top_down_{l}"] = [S_p_td, S_m_td]
            connections[f"top_down_2_{l}"] = [S_p_td_2, S_m_td_2]
            
            
    # Os -> Es
    S_o_external_p, S_o_external_m = make_top_down_connections(Os, Es_0[-1], Es_1[-1])
    S_o_external_2_p, S_o_external_2_m = make_top_down_connections_2(Es_0[-1], Es_1[-1], Os)

            
    # Here we are trying to find a good value to init connections with the output layer.        
    S_o_external_p.connect()
    S_o_external_p.w = 'abs(randn()) * sigma_w_init'
    S_o_external_p.connect()
    S_o_external_p.w = 'abs(randn()) * sigma_w_init'
    
    S_o_external_2_m.connect()
    S_o_external_2_m.w = 'abs(randn()) * sigma_w_init'
    S_o_external_2_m.connect()
    S_o_external_2_m.w = 'abs(randn()) * sigma_w_init'
    
   
    connections["output_external_p"] = [S_o_external_p]
    connections["output_external_m"] = [S_o_external_m]
    
    connections["output_external_2_p"] = [S_o_external_2_p]
    connections["output_external_2_m"] = [S_o_external_2_m]
    
    

    S_gist_input, S_gist_output = make_gist_connections(Rs, G)
    connections["gist_input"] = [S_gist_input]
    connections["gist_output"] = [S_gist_output]

    # External output connections
    # In original code: attempted to connect Rs[-1] to Os
    # We'll do a single large synapse group for simplicity:


  
    # return Rs, Es_0, Es_1, G, Os, S_o_internal, connections
    
    return Rs, Es_0, Es_1, G, Os,  connections





# Used to normalize the 2d gaussian processes (curves) which make up the dataset, in order to provide a meaningful value range
# for the input neurons
def normalize_tensor(tensor, old_min, old_max, new_min, new_max):
    normalized_tensor = (tensor - old_min) / (old_max - old_min)
    scaled_tensor = normalized_tensor * (new_max - new_min) + new_min
    return scaled_tensor



# This returns an easy-to-read, normalized gaussian process dataset
def normalize_and_unwrap_dataset(dt, minval, maxval):
    minimum = 0
    maximum = 0
    all_curves = [torch.tensor(curve, dtype=torch.float64) for curve in chain.from_iterable(dt["curves"].values())]
    for curve in all_curves:
        _min = torch.min(curve)
        _max = torch.max(curve)
        if (minimum > _min): minimum = _min
        if (maximum < _max): maximum = _max

    unwrapped = []
    for key in dt["categories"].keys():
        for curve in dt["curves"][key]:
            unwrapped.append((torch.tensor(key), normalize_tensor(torch.tensor(curve), minimum, maximum, minval, maxval)))
    return unwrapped


# This function converts a class label of a gaussian-process-curve of our dataset
# into the indexes of the neurons of the output layer which need to be stimulated.
# For example, if the class label is (1,0,2) and num_classes_per_layer = 3 the stimulated neurons are 1, 3 + 1 = 4, 3^2 + 2 = 11.
def get_output_stimuli_indexes_from_labels(labels, num_classes_per_layer):
    if labels.ndim == 1:
        labels = labels.unsqueeze(0)

    indexes = torch.zeros_like(labels)
    for ii in range(labels.shape[1]):
        summed = 0
        if ii > 0:
            summed = torch.stack([labels[:, j]*num_classes_per_layer**(ii-j) for j in range(ii)]).sum(dim=0)
        indexes[:, ii] = summed + labels[:, ii]

    # Adjust indexes for concatenation
    to_sum = 0
    for ii in range(labels.shape[1]):
        if ii > 0:
            to_sum += num_classes_per_layer**ii
        indexes[:, ii] = indexes[:, ii] + to_sum
    return indexes


# From neuroin indexes to current array
def get_output_current_arrays(I_indexes, dim, I_value):
    if I_indexes.ndim == 1:
        I_indexes = I_indexes.unsqueeze(0)

    out = torch.zeros((I_indexes.shape[0], dim), dtype=torch.float32)
    row_indices = torch.arange(I_indexes.shape[0]).repeat_interleave(I_indexes.shape[1])
    col_indices = I_indexes.flatten()
    out[row_indices, col_indices] = I_value
    return out


# pythorch-like dataset function to iterate through our hyerarchical gaussian process dataset.
class CurveDataset(Dataset):
    def __init__(self, data, minval, maxval):
        self.data = normalize_and_unwrap_dataset(data, minval, maxval)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        labels, curves = self.data[idx]
        return labels, curves.flatten()
        
# This gets called by the dataloader before outputting each batch. It transforms curves and labels into TimedArrays, which are seen from brian2
# As current inputs to input and output neurons.
def collate_fn(batch, msec_step, num_pause_blocks, num_classes_per_layer, out_dim):
    labels, curves = zip(*batch)
    labels = torch.stack(labels, dim=0)
    curves = torch.stack(curves, dim=0)

    pause = torch.zeros_like(curves[0])
    pause_block = torch.tile(pause, (num_pause_blocks, 1))
    curves_with_pause = torch.vstack([torch.vstack((row, pause_block)) for row in curves])
    visual_stimulus = TimedArray(curves_with_pause.numpy() * nA, dt=msec_step)

    stimulus_indexes = TimedArray(np.tile([1] + [0] * num_pause_blocks, curves.size(0)), dt=msec_step)

    out_stimuli_idx = get_output_stimuli_indexes_from_labels(labels, num_classes_per_layer)
    out_stimuli_array = get_output_current_arrays(out_stimuli_idx, out_dim, 1)
    label_pause = torch.zeros(out_dim)
    label_pause_block = torch.tile(label_pause, (num_pause_blocks, 1))
    
    out_stimuli_with_pause = torch.vstack([torch.vstack((row, label_pause_block)) for row in out_stimuli_array])
    output_stimuli = TimedArray(out_stimuli_with_pause.numpy() * nA, dt=msec_step)
    
    #out_stimuli_with_pause = np.zeros_like(torch.vstack([torch.vstack((row, label_pause_block)) for row in out_stimuli_array]))
    #output_stimuli = TimedArray(out_stimuli_with_pause * nA, dt=msec_step)


    return stimulus_indexes, output_stimuli, visual_stimulus

###############################################################################
# Build the network and run
###############################################################################
#dims = [22*22, 250, 125, 64]
N = 22
dims = [N*N,20,20,20]
num_classes_per_layer = 3
max_depth = 3

# Actually create the SNN
# Rs, Es_0, Es_1, G, Os, S_o_internal, connections = make_network(dims, num_classes_per_layer, max_depth)
Rs, Es_0, Es_1, G, Os, connections = make_network(dims, num_classes_per_layer, max_depth)

Rs[0].indices = range(Rs[0].N)

# Create the dataset
su = Synthetic_Dataset_Utils()
ranges = [30, 30, 30, 30]
prior_params = [10, 10, 10, 10]
num_samples_per_class = 500

tree = su.build_tree(prior_params, 0, max_depth, num_classes_per_layer, 1, ranges)
synth_dataset = su.make_dataset(tree, num_classes_per_layer, num_samples_per_class=num_samples_per_class, N=N)

output_dim = Os.N
# lr = 0.03
lr = 0.0002
batch_size = 40
minibatch_size = 5
num_pause_blocks = 1

dataset = CurveDataset(synth_dataset, 0.3, 1.2)
dl = DataLoader(
    dataset,
    shuffle=True,
    batch_size=batch_size,
    collate_fn=partial(
        collate_fn,
        msec_step=msec_step,
        num_pause_blocks=num_pause_blocks,
        num_classes_per_layer=num_classes_per_layer,
        out_dim=output_dim
    )
)



torch.save(dataset, "data/dataset1.pt")

#dataset = torch.load("data/dataset1.pt")




@network_operation(dt=msec_step, when='end')
def debug():
    pass
    #print("Rs[0].I: ", Rs[0].I[:])
    
    
def update_batch_sum(synapses):
    synapses.batch_sum_X += synapses.X  # Perform the operation
    
# In order to be able to update the weights in batches, we keep track of the cumulative value of X over a minibatch   
@network_operation(dt=msec_step, when='end')
def update_sums():
    current_idx = int(defaultclock.t / msec_step)
    if stimulus_indexes.values[current_idx] == 1:
        for n in range(len(Rs)):
            if n > 0:
                Rs[n].batch_sum_X += Rs[n].X
            Es_0[n].batch_sum_X += Es_0[n].X
            Es_1[n].batch_sum_X += Es_1[n].X
            
    Os.batch_sum_X += Os.X
            

# Used to regularize only non-zero, positive weights
def if_positive(x, w):
    # Return x where w > 0, otherwise 0 (element-wise)
    return np.where(w > 0, x, 0 * volt)



# This important function implements hebbian learning. It updates the weigght every msec_step*minibatch_size
debug_weights = []
@network_operation(dt=msec_step*minibatch_size, when='end')
def apply_weight_update():
    print("Applying weight update")
    for n in range(len(Rs)-1):
        
        print("n: ", n)
        
        # Retrieve top_down synapses       
        S_p, S_m = connections[f"top_down_{n}"]
        S_p_2, S_m_2 = connections[f"top_down_2_{n}"]
        
        # Multiply the (unitless) accumulated X of each synapse pre and post synaptic neurons, and use that to determine the update. If the weight
        # is positive, it tends to decrease (regularizer).
        for j_idx in range(Es_0[n].N):
            for i_idx in range(Rs[n+1].N):
                
                positive_connections_delta = lr * Es_0[n].batch_sum_X_[j_idx] * Rs[n+1].batch_sum_X_[i_idx] - if_positive(regularizer, S_p.w_[i_idx, j_idx])
                negative_connections_delta = lr * Es_1[n].batch_sum_X_[j_idx] * Rs[n+1].batch_sum_X_[i_idx] - if_positive(regularizer, S_m.w_[i_idx, j_idx])
                
                S_p.w[i_idx, j_idx] += positive_connections_delta/minibatch_size  * siemens
                S_m.w[i_idx, j_idx] -= negative_connections_delta/minibatch_size * siemens
                
                S_p_2.w[i_idx, j_idx] += positive_connections_delta/minibatch_size * siemens
                S_m_2.w[i_idx, j_idx] -= negative_connections_delta/minibatch_size * siemens
                
                
                    
            
            # Reset accumulated X after update
            Rs[n+1].batch_sum_X[:] = 0 * volt
            Es_0[n].batch_sum_X[:] = 0 * volt
            Es_1[n].batch_sum_X[:] = 0 * volt

        # Debug
        print("hehd: ", n+1)
        if (n+1) == 1:
            print("iuhichseiuhsi: ", positive_connections_delta)
            print("iuhichsesfei: ", negative_connections_delta)
            debug_weights.append(S_p.w)
    
    
    
    external_syn_p = connections["output_external_p"][0]
    external_syn_2_p = connections["output_external_2_p"][0]
    
    external_syn_m = connections["output_external_m"][0]
    external_syn_2_m = connections["output_external_2_m"][0]
    
    
    # Update synapses between last error layer and Output layer
    for i_idx in range(Es_0[-1].N):
        for j_idx in range(Os.N):
            
            delta_p = lr * Es_0[-1].batch_sum_X_[i_idx] * Os.batch_sum_X_[j_idx] * siemens
            external_syn_p.w[i_idx, j_idx] += delta_p/minibatch_size
            external_syn_2_p.w[i_idx, j_idx] += delta_p/minibatch_size
            
            delta_m = lr * Es_1[-1].batch_sum_X_[i_idx] * Os.batch_sum_X_[j_idx] * siemens
            external_syn_m.w[i_idx, j_idx] += delta_m/minibatch_size
            external_syn_2_m.w[i_idx, j_idx] += delta_m/minibatch_size
            
            
    Es_0[-1].batch_sum_X[:] = 0
    Es_1[-1].batch_sum_X[:] = 0
    Os.batch_sum_X[:] = 0
    
    # Automatic non-negativity for all Synapses objects
    #print("CHUNGUS: ")
    for obj in net.objects:
        if isinstance(obj, Synapses):
            # Zero out negative weights
            #print("Chungus")
            #print("A")
            #print(obj.w)
            #print("Negative weights:")
            #print(obj.w[obj.w < 0 * nsiemens])
            obj.w_[:] = np.clip(obj.w_[:], 0, None)
            #print("B")
            #print(obj.w)
        
            
# The following is to be sure to construct a functioning Network object            
     
# Extract all synapses and objects for the network
all_synapses = []
# all_synapses.append(S_o_internal)

for ii in range(len(Rs)-1):
    bu_syns = connections[f"bottom_up_{ii}"]
    all_synapses.extend(bu_syns)
    td_key = f"top_down_{ii}"
    if td_key in connections:
        td_syns = connections[td_key]
        all_synapses.extend(td_syns)

for ii in range(1,len(Rs)-1):
    bu_syns = connections[f"bottom_up_2_{ii}"]
    all_synapses.extend(bu_syns)
    td_key = f"top_down_2_{ii}"
    if td_key in connections:
        td_syns = connections[td_key]
        all_synapses.extend(td_syns)

gist_input_syn = connections["gist_input"][0]
gist_output_syn = connections["gist_output"][0]
all_synapses.append(gist_input_syn)
all_synapses.append(gist_output_syn)

external_syn_p = connections["output_external_p"][0]
external_syn_m = connections["output_external_m"][0]

external_syn_2_p = connections["output_external_2_p"][0]
external_syn_2_m = connections["output_external_2_m"][0]

all_synapses.append(external_syn_p)
all_synapses.append(external_syn_m)

all_synapses.append(external_syn_2_p)
all_synapses.append(external_syn_2_m)

monitor_Es1_a = StateMonitor(Es_1[0], ['v'], record=[0])
monitor_Es1_b = StateMonitor(Es_1[0], ['I_tot'], record=[0])
#spike_monitor = SpikeMonitor(Rs[0])

s = 0

        



monitor_o = StateMonitor(Os, ['v', 'I_tot', 'X'], record=True)
spike_mon_o = NamedSpikeMonitor(Os, record = True, custom_name = "output")

'''
net = Network(
    Rs, Es_0, Es_1, G, Os,
    all_synapses,        
    monitor_o, spike_mon_o,         # <-- Add these
    debug, update_sums, apply_weight_update
)
'''

# Add spike monitors for all neuron groups

spike_mon_G = NamedSpikeMonitor(G, record=True, custom_name = "Gist")

stimulus_indexes, output_stimuli, input_stimuli = next(iter(dl))

debug_monitor = NamedSpikeMonitor(Es_0[1], record=True, custom_name = f"Es_0_[1]_debug")
net = Network(
    Rs, Es_0, Es_1, G, Os,  # Neuron groups       
    monitor_o, spike_mon_o, debug_monitor,monitor_Es1_a,monitor_Es1_b, # Output monitors
    spike_mon_G,  # Spike monitors
    debug, update_sums, apply_weight_update # Additional operations
)

for l in range(len(Rs)):
    net.add(NamedSpikeMonitor(Rs[l], record=True, custom_name = f"R[{l}]"))
   
    net.add(NamedSpikeMonitor(Es_0[l], record=True, custom_name = f"Es_0[{l}]"))
    net.add(NamedSpikeMonitor(Es_1[l], record=True, custom_name = f"Es_1[{l}]"))

for synapse in all_synapses:
    net.add(synapse)

all_batch_rates = []
output_spikes = []

s = 0

#stimulus_indexes, output_stimuli, input_stimuli = next(iter(dl))

#net.stimulus_indexes = stimulus_indexes
#net.output_stimuli   = output_stimuli
#net.input_stimuli    = input_stimuli

run_time = (batch_size + batch_size*num_pause_blocks)*100*ms
net.run(run_time)
result_dir = f'results_v2'
#device.build(run=True, directory="project_v2",results_directory=result_dir) 

# device.build(directory='project', run = False)

states = net.get_states()
print("FUUUUUUUUUUUUCK")
b2t.brian_plot(monitor_Es1_a)
plt.show()
b2t.brian_plot(monitor_Es1_b)
plt.show()
keys = []
for key in states.keys():
    if "spike" in key:
        keys.append(key)
        print(key)
        
for key in keys:
    spike_mon_o = net[key]
    spikes = spike_mon_o.spike_trains()
    
    b2t.brian_plot(spike_mon_o)
    # Plotting the spike train with dots
    plt.figure(figsize=(10, 6))
    ii=0
    old = 0
    for neuron, spike_times in spikes.items():
    
        old = spike_times
        
        if ii < 100:
            plt.scatter(spike_times/ms, [neuron] * len(spike_times), s=10, label=f'Neuron {neuron}' if neuron == 0 else "")
            ii+=1

    plt.xlabel('Time (ms)')
    plt.ylabel('Neuron index')
    plt.title(f'{spike_mon_o.custom_name}. Num neurons = {len(spikes.items())}')
    print(spike_mon_o.source)
    plt.yticks(range(N))
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
    
```

```python
# Parameters for neurons
Cm = 281 * pF  # Membrane capacitance
gL = 30 * nS   # Leak conductance
EL = -70.6 * mV  # Leak reversal potential
Vth = -50.4 * mV  # Spike threshold
DeltaT = 2 * mV  # Slope factor
Vr = -70.6 * mV  # Reset potential
Vcut = -40 * mV  # Cutoff potential for spike generation
tau_A = 1 * ms  # Adaptation time constant
c = 4 * nS       # Coupling parameter
b = 0.0805 * nA  # Spike-triggered adaptation increment

# Parameters for synapses
tau_rise = 5 * ms  # Rise time constant for AMPA
tau_decay = 50 * ms  # Decay time constant for NMDA
w_init = 0.5 * nS  # Initial synaptic weight (conductance)

# LSD-modulated parameters
DeltaT_LSD = 3 * mV
gL_LSD = 20 * nS
Vth_LSD = -52 * mV
b_LSD = 0.06 * nA
tau_A_LSD = 14.05 * ms


import os

import time
import torch
import datetime
import brian2 as b2
from brian2 import *
from itertools import chain
from functools import partial
import matplotlib.pyplot as plt

from torch.utils.data import Dataset, DataLoader
import numpy as np

from dataset_util import Synthetic_Dataset_Utils




from brian2 import prefs
#from brian2 import set_device
from brian2.devices import device_module

import brian2cuda
from brian2cuda import *

import brian2tools as b2t

from brian2 import prefs


prefs.codegen.target = 'cython'



#set_device("cpp_standalone", directory='project', build_on_run=False)

#device = b2.get_device()


#set_device("cuda_standalone", build_on_run=True)
# Set specific preferences for CUDA standalone


prefs.devices.cuda_standalone.cuda_backend.cuda_runtime_version = (12.2)
prefs.devices.cuda_standalone.profile_statemonitor_copy_to_host = None
prefs.devices.cuda_standalone.cuda_backend.extra_compile_args_nvcc = ['-use_fast_math', '-w']
prefs.devices.cuda_standalone.cuda_backend.detect_gpus = True
prefs.devices.cuda_standalone.cuda_backend.gpu_id = 0  # Use the first GPU
# prefs.devices.cuda_standalone.cuda_backend.cuda_path = '/home/brovoletto/SerotoninSNN/nvcc_installation/nvcc_folder'



# REMEMBER YOU SWITCHED SOME EXCITATORY WITH INHIBITORY!!!!!!!!


###############################################################################
# Model Parameters and Equations
###############################################################################
msec_step = 400 * ms

# TODO: Make weights always non-negative!


# Equations for everything. I_post in the synapse is modifying I in the target neuron


# TODO: Make weights always non-negative! (maybe not necessary)
# Equations for everything. I_post in the synapse is modifying I in the target neuron



# Parameters for neurons
Cm = 281 * pF  # Membrane capacitance
gL = 30 * nS   # Leak conductance
EL = -70.6 * mV  # Leak reversal potential
Vth = -45.4 * mV  # Spike threshold
Vth_out = -25.4 * mV  # Spike threshold

DeltaT = 5 * mV  # Slope factor
Vr = -70.6 * mV  # Reset potential

Vcut = -40 * mV  # Cutoff potential for spike generation
tau_A = 300 * ms  # Adaptation time constant
c = 10 * nS       # Coupling parameter
b = 2 * nA  # Spike-triggered adaptation increment

# Parameters for synapses
tau_rise = 5 * ms  # Rise time constant for AMPA
tau_decay = 50 * ms  # Decay time constant for NMDA
sigma_w_init = 3 * nsiemens   # Initial synaptic weight (conductance)
w_init_2 = 3 * nsiemens

regularizer = 10 * nvolt



#-----------------------------

# Parameters for neurons
Cm = 281 * pF  # Membrane capacitance
gL = 30 * nS   # Leak conductance
EL = -70.6 * mV  # Leak reversal potential
Vth = -45.4 * mV  # Spike threshold
DeltaT = 2 * mV  # Slope factor
Vr = -70.6 * mV  # Reset potential
Vcut = -40 * mV  # Cutoff potential for spike generation
tau_A = 1 * ms  # Adaptation time constant
c = 4 * nS       # Coupling parameter
b = 1 * nA  # Spike-triggered adaptation increment

# Parameters for synapses
tau_rise = 5 * ms  # Rise time constant for AMPA
tau_decay = 50 * ms  # Decay time constant for NMDA
w_init = 0.5 * nS  # Initial synaptic weight (conductance)

# LSD-modulated parameters
DeltaT = 3 * mV
gL= 20 * nS
Vth = -52 * mV
b = 1 * nA
tau_A = 14.05 * ms



# AdEx neuron equations. Separate terms are used for each I contribution, in order to work in brian2. Inhibitory currents are inhibitory
# simply by means of the minus sign.

# Spike trace terms (X and Y) are implemented following the equetions described in "Predictive coding with spiking neurons and feedforward gist signaling".
eqs = '''
dv/dt = ( -gL * (v - EL) + gL * DeltaT * exp((v - Vth) / DeltaT) + I_tot - A ) / Cm : volt
dA/dt = (c * (v - EL) - A) / tau_A : amp
dX/dt = -X / tau_decay + Y / tau_rise : volt    # Spike trace
dY/dt = -Y / tau_decay : volt  # Glutamate decay
Iexc : amp
Iexc_2 : amp
Iexc_3 : amp
Iexc_4 : amp
Iinh : amp
Iinh_2 : amp
Iinh_3 : amp
Iinh_4 : amp
I_tot = Iexc + Iexc_2 + Iexc_3 + Iexc_4 - Iinh - Iinh_2 - Iinh_3 - Iinh_4 : amp
batch_sum_X : volt   # Accumulate X values over a batch

'''
# Ii = I + input_stimuli(t)  :  amp  # Input current

# These take inputs from the input_stimuli TimedArray
input_eqs = '''
dv/dt = ( -gL * (v - EL) + gL * DeltaT * exp((v - Vth) / DeltaT) + I - A ) / Cm : volt
dA/dt = (c * (v - EL) - A) / tau_A : amp
dX/dt = -X / tau_decay + Y / tau_rise : volt    # Spike trace
dY/dt = -Y / tau_decay : volt  # Glutamate decay
I = input_stimuli(t, i) : amp
indices : integer
'''

# These take inputs from output_stimuli TimedArray
out_eqs = '''
dv/dt = ( -gL * (v - EL) + gL * DeltaT * exp((v - Vth) / DeltaT) + I_tot - A ) / Cm : volt
dA/dt = (c * (v - EL) - A) / tau_A : amp
dX/dt = -X / tau_decay + Y / tau_rise : volt    # Spike trace
dY/dt = -Y / tau_decay : volt  # Glutamate decay
Iexc_internal : amp  # Synaptic input
Iexc_3 : amp
Iexc_4 : amp
Iinh_3 : amp
Iinh_4 : amp
I_tot = Iexc_3 + Iexc_4 - Iinh_3 - Iinh_4 + Iexc_internal + output_stimuli(t, i) : amp
indices : integer  # dimensionless index variable
spike_count = 0 : integer
batch_sum_X : volt
'''

# Synapse equations (Spike trace dynamics)

# Each synapse has a weight parameter, which determines how much its spike trace X influences the current I of the postsynaptic neuron.
syn_eqs_exc = '''
w : siemens   # Synaptic weight, conductance

Iexc_post = w * X_pre : amp (summed)
'''

syn_eqs_exc_2 = '''
w : siemens   # Synaptic weight, conductance

Iexc_2_post = w * X_pre : amp (summed)
'''

syn_eqs_exc_3 = '''
w : siemens

Iexc_3_post = w * X_pre : amp (summed)
'''

syn_eqs_exc_4 = '''
w : siemens

Iexc_4_post = w * X_pre : amp (summed)
'''


syn_eqs_exc_output = '''
w : siemens   # Synaptic weight, conductance

Iexc_internal_post = w * X_pre  : amp (summed)
'''



syn_eqs_inh= '''
w : siemens   # Synaptic weight, conductance

Iinh_post = w * X_pre  : amp (summed)
'''

syn_eqs_inh_2= '''
w : siemens   # Synaptic weight, conductance

Iinh_2_post = w * X_pre  : amp (summed)
'''

syn_eqs_inh_3= '''
w : siemens   # Synaptic weight, conductance

Iinh_3_post = w * X_pre : amp (summed)
'''

syn_eqs_inh_4= '''
w : siemens   # Synaptic weight, conductance

Iinh_4_post = w * X_pre  : amp (summed)
'''


# As in the paper, when a neuron spikes it causes an immediate change in the value of Y and A.

on_pre_value = 'Y_post = 230*mvolt; A_post = A_post + b'


# This function takes a dims list and makes len(dims) neuron groups, of the sizes specified in dims
def make_groups(dims, eqs=eqs, with_input=False):
    groups = []
    for ii, dim in enumerate(dims):
        if ii == 0 and with_input:
            g = NeuronGroup(dim, input_eqs, threshold='v > -40*mV', reset='v = Vr', method='euler')
        else:
            g = NeuronGroup(dim, eqs, threshold='v > -40*mV', reset='v = Vr', method='euler')
        g.v = EL
        groups.append(g)
    return groups


# We add a custom_name attribute to synapses for debugging purposes, in order to identify them more easily since brian2 changes their name.
class NamedSynapses(Synapses):
    def __init__(self, source, target=None, model=None, custom_name=None, **kwargs):
        super().__init__(source, target, model, **kwargs)
        
        self.add_attribute("brian2_name")
        self.add_attribute("custom_name")
        
        self.brian2_name = self.name  
        
        self.custom_name = custom_name or f"{source}-{target}"
        
        print(self.brian2_name, self.custom_name)
        
    def connect(self, **kwargs):
        # Call the original connect method
        super().connect(**kwargs)
        if self.custom_name == "suygu5":
            # Print the connections
            connections = [(ii, j) for ii, j in zip(self.i[:], self.j[:])]
            print("Connections created:")
            for pre, post in connections:
                print(f"Pre: {pre}, Post: {post}")
            
        
# Same for the monitors
class NamedSpikeMonitor(SpikeMonitor):
    def __init__(self, group, custom_name=None, **kwargs):
        super().__init__(group, **kwargs)
        self.add_attribute("custom_name")
        self.custom_name = custom_name or str(group)    
     
     
# This function can be useful when using brian2cuda since .connect() may not work correctly. Not used here.
def manual_fully_connect(syn, N_pre, N_post):
    # Manually specify connections
    pre_indices = [ii for ii in range(N_pre) for _ in range(N_post)]
    post_indices = [j for _ in range(N_pre) for j in range(N_post)]

    # Connect manually
    syn.connect(i=pre_indices, j=post_indices) 
    
    return syn  
            

# This function is called to make bottom-up connections. They are one-to-one (not fully connected) and they do not evolve over time.
# We are following "Predictive coding with spiking neurons and feedforward gist signaling".
# They connect the representation unit Rs[n] with the error units Es_0[n] (excitatory) and Es_1[n] (inhibitory) of the same area
def make_bottom_up_connections(R, E_0, E_1):
    print("A")
    print("Bottom-up: ", R.N, E_0.N, E_1.N )
    S_p = NamedSynapses(R, E_0, model=syn_eqs_exc, on_pre= on_pre_value, custom_name = "s1")
    S_p.connect(j='i')
    S_p.w = 'w_init_2'

    S_m = NamedSynapses(R, E_1, model=syn_eqs_inh, on_pre= on_pre_value, custom_name = "s2")
    S_m.connect(j='i')
    S_m.w = 'w_init_2'

    return S_p, S_m



# Name is slightly misleading. They are one to one connections paired with the ones from the function above, but signal in the opposite direction.
def make_bottom_up_connections_2(E_0, E_1,R):
    print("B")
    print("Bottom-up-2: ", R.N, E_0.N, E_1.N )
    S_p = NamedSynapses(E_0, R, model=syn_eqs_inh_2, on_pre= on_pre_value, custom_name = "s3")
    S_p.connect(j='i')
    S_p.w = 'w_init_2'

    S_m = NamedSynapses(E_1, R,  model=syn_eqs_exc_2, on_pre= on_pre_value, custom_name = "s4")
    S_m.connect(j='i')
    S_m.w = 'w_init_2'

    return S_p, S_m


# This function makes fully connected synapses which then evolve during training. 
# They connect the error units Es_0[n] (excitatory) and Es_1[n] (inhibitory) with the representation unit Rs[n+1] of the area above.
# abs(randn()) creates a half gaussian with only positive values.

def make_top_down_connections(R, E_0, E_1):
    print("C")
    print("Top-down: ", R.N, E_0.N, E_1.N )
    
    S_p = NamedSynapses(R, E_0, model=syn_eqs_inh_3, on_pre= on_pre_value, custom_name = "s6")
    S_p.connect()
    #manual_fully_connect(S_m, R.N, E_0.N)
    S_p.w = 'abs(randn()) * sigma_w_init'
    
    
    S_m = NamedSynapses(R, E_1, model=syn_eqs_exc_3, on_pre= on_pre_value, custom_name = "s5")
    S_m.connect()
    #manual_fully_connect(S_p, R.N, E_1.N)
    S_m.w = 'abs(randn()) * sigma_w_init'



    return S_p, S_m


# As the function above, but in the opposite direction.
def make_top_down_connections_2(E_0, E_1, R):
    print("D")
    print("Top-down-2: ", R.N, E_0.N, E_1.N )
    
    S_p_2 = NamedSynapses(E_0, R,  model=syn_eqs_exc_4, on_pre= on_pre_value, custom_name = "s8")
    S_p_2.connect()
    #manual_fully_connect(S_m_2, E_0.N, R.N)
    S_p_2.w = 'abs(randn()) * sigma_w_init'
    
    
    S_m_2 = NamedSynapses(E_1, R,  model=syn_eqs_inh_4, on_pre= on_pre_value, custom_name = "s7")
    S_m_2.connect()
    #manual_fully_connect(S_p_2, E_1.N, R.N)
    S_m_2.w = 'abs(randn()) * sigma_w_init'



    return S_p_2, S_m_2


# Gist connections as in the paper. They connect Rs[0] with every other representation unit. Weights are fixed, not learned.
def make_gist_connections(Rs, G):
    S_gist_input = NamedSynapses(Rs[0], G, model=syn_eqs_exc, on_pre= on_pre_value, custom_name = "s9")
    S_gist_input.connect(p=0.05)
    S_gist_input.w = 'abs(randn()) * sigma_w_init * 2'

    # Connect G to all other layers
    S_gist_output = NamedSynapses(G, Rs[-1], model=syn_eqs_exc, on_pre= on_pre_value, custom_name = "s10")
    S_gist_output.connect(p=0.05)
    S_gist_output.w = 'abs(randn()) * sigma_w_init * 2'
    return S_gist_input, S_gist_output


# The output layer mirrors the possible hierarchical classes in the dataset. Therefore, if the dataset-creation parameters num_classes_per_layer = 3 and max_depth = 4 
# Then, the output layer is going to mirror the dataset's classes, by having 4 layers of size [3, 3^2, 3^3, 3^4]. The reason is that each child class
# produces a number of child classes equal to num_classes_per_layer.

def make_output_layer(num_classes_per_layer, max_depth):
    
    '''
    if w_out_init is None:
        w_out_init = w_init/3000
        print("W_OUT_INIT: ", w_out_init)
    '''

    total_neurons = sum([num_classes_per_layer**(d) for d in range(max_depth)])
    Os = NeuronGroup(N=total_neurons, model=out_eqs, threshold='v > -40*mV', reset='v = Vr', method='euler')

    #S_o = NamedSynapses(Os, Os, model=syn_eqs_exc_output, on_pre= on_pre_value)

    trace = 0
    for d in range(num_classes_per_layer - 1):
        new_trace = trace + num_classes_per_layer**(d+1)
        source_indices = list(range(trace, trace + num_classes_per_layer**(d+1)))
        target_indices = list(range(new_trace, new_trace + num_classes_per_layer**(d+2)))
        ii, jj = np.meshgrid(source_indices, target_indices, indexing='ij')
        #S_o.connect(i=ii.flatten(), j=jj.flatten())
        trace = new_trace

    # Assign indices for output stimuli
    stimulus_indices = []
    for ii in range(max_depth):
        for _ in range(num_classes_per_layer**(ii+1)):
            stimulus_indices.append(ii)
    Os.indices = stimulus_indices

    return Os#, S_o


# This function uses the previously defined functions to construct the whole network. len(dims) determines the depth of the network, and the
# values inside dims determine the size of each layer. Each value corresponds to the size of Rs[n] and Es_0[n] and Es_1[n].

def make_network(dims, num_classes_per_layer, max_depth):
    max_depth += 1

    # Using dims, we make representation layers and corresponding excitatory (0) and inhibitory (1) Error layers.
    Rs = make_groups(dims, with_input=True)
    Es_0 = make_groups(dims)
    Es_1 = make_groups(dims)
    # Gist neurons, 16 as in the paper
    G = NeuronGroup(16, eqs, threshold='v > -40*mV', reset='v = Vr', method='euler')

    # Optional synapses inside the outpt layer, to connect parent neurons to child neurons.
    #Os, S_o_internal = make_output_layer(num_classes_per_layer, max_depth, w_out_init=w_out_init)
    Os = make_output_layer(num_classes_per_layer, max_depth)
    #w_out_init = w_init/100
    
    connections = {}
    for l in range(len(Rs)):
        
        S_p, S_m = make_bottom_up_connections(Rs[l], Es_0[l], Es_1[l])
        connections[f"bottom_up_{l}"] = [S_p, S_m]

        # Backward one-to-one connections are NOT created from Rs[1] to Rs[0] (the input layer)
        if l != 0:
            
            
            S_p_2, S_m_2 = make_bottom_up_connections_2(Es_0[l], Es_1[l], Rs[l])
            connections[f"bottom_up_2_{l}"] = [S_p_2, S_m_2]
        
        # top down connections between representation unit and error units of the previous layer. 
        # if l = len(Rs), the error units need to connect to the output layer (Os). This happens below.
        if l != len(Rs)-1:

            S_p_td, S_m_td = make_top_down_connections(Rs[l+1], Es_0[l], Es_1[l])
            S_p_td_2, S_m_td_2 = make_top_down_connections_2(Es_0[l], Es_1[l], Rs[l+1])
            connections[f"top_down_{l}"] = [S_p_td, S_m_td]
            connections[f"top_down_2_{l}"] = [S_p_td_2, S_m_td_2]
            
            
    # Os -> Es
    S_o_external_p, S_o_external_m = make_top_down_connections(Os, Es_0[-1], Es_1[-1])
    S_o_external_2_p, S_o_external_2_m = make_top_down_connections_2(Es_0[-1], Es_1[-1], Os)

            
    # Here we are trying to find a good value to init connections with the output layer.        
    S_o_external_p.connect()
    S_o_external_p.w = 'abs(randn()) * sigma_w_init'
    S_o_external_p.connect()
    S_o_external_p.w = 'abs(randn()) * sigma_w_init'
    
    S_o_external_2_m.connect()
    S_o_external_2_m.w = 'abs(randn()) * sigma_w_init'
    S_o_external_2_m.connect()
    S_o_external_2_m.w = 'abs(randn()) * sigma_w_init'
    
   
    connections["output_external_p"] = [S_o_external_p]
    connections["output_external_m"] = [S_o_external_m]
    
    connections["output_external_2_p"] = [S_o_external_2_p]
    connections["output_external_2_m"] = [S_o_external_2_m]
    
    

    S_gist_input, S_gist_output = make_gist_connections(Rs, G)
    connections["gist_input"] = [S_gist_input]
    connections["gist_output"] = [S_gist_output]

    # External output connections
    # In original code: attempted to connect Rs[-1] to Os
    # We'll do a single large synapse group for simplicity:


  
    # return Rs, Es_0, Es_1, G, Os, S_o_internal, connections
    
    return Rs, Es_0, Es_1, G, Os,  connections





# Used to normalize the 2d gaussian processes (curves) which make up the dataset, in order to provide a meaningful value range
# for the input neurons
def normalize_tensor(tensor, old_min, old_max, new_min, new_max):
    normalized_tensor = (tensor - old_min) / (old_max - old_min)
    scaled_tensor = normalized_tensor * (new_max - new_min) + new_min
    return scaled_tensor



# This returns an easy-to-read, normalized gaussian process dataset
def normalize_and_unwrap_dataset(dt, minval, maxval):
    minimum = 0
    maximum = 0
    all_curves = [torch.tensor(curve, dtype=torch.float64) for curve in chain.from_iterable(dt["curves"].values())]
    for curve in all_curves:
        _min = torch.min(curve)
        _max = torch.max(curve)
        if (minimum > _min): minimum = _min
        if (maximum < _max): maximum = _max

    unwrapped = []
    for key in dt["categories"].keys():
        for curve in dt["curves"][key]:
            unwrapped.append((torch.tensor(key), normalize_tensor(torch.tensor(curve), minimum, maximum, minval, maxval)))
    return unwrapped


# This function converts a class label of a gaussian-process-curve of our dataset
# into the indexes of the neurons of the output layer which need to be stimulated.
# For example, if the class label is (1,0,2) and num_classes_per_layer = 3 the stimulated neurons are 1, 3 + 1 = 4, 3^2 + 2 = 11.
def get_output_stimuli_indexes_from_labels(labels, num_classes_per_layer):
    if labels.ndim == 1:
        labels = labels.unsqueeze(0)

    indexes = torch.zeros_like(labels)
    for ii in range(labels.shape[1]):
        summed = 0
        if ii > 0:
            summed = torch.stack([labels[:, j]*num_classes_per_layer**(ii-j) for j in range(ii)]).sum(dim=0)
        indexes[:, ii] = summed + labels[:, ii]

    # Adjust indexes for concatenation
    to_sum = 0
    for ii in range(labels.shape[1]):
        if ii > 0:
            to_sum += num_classes_per_layer**ii
        indexes[:, ii] = indexes[:, ii] + to_sum
    return indexes


# From neuroin indexes to current array
def get_output_current_arrays(I_indexes, dim, I_value):
    if I_indexes.ndim == 1:
        I_indexes = I_indexes.unsqueeze(0)

    out = torch.zeros((I_indexes.shape[0], dim), dtype=torch.float32)
    row_indices = torch.arange(I_indexes.shape[0]).repeat_interleave(I_indexes.shape[1])
    col_indices = I_indexes.flatten()
    out[row_indices, col_indices] = I_value
    return out


# pythorch-like dataset function to iterate through our hyerarchical gaussian process dataset.
class CurveDataset(Dataset):
    def __init__(self, data, minval, maxval):
        self.data = normalize_and_unwrap_dataset(data, minval, maxval)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        labels, curves = self.data[idx]
        return labels, curves.flatten()
        
# This gets called by the dataloader before outputting each batch. It transforms curves and labels into TimedArrays, which are seen from brian2
# As current inputs to input and output neurons.
def collate_fn(batch, msec_step, num_pause_blocks, num_classes_per_layer, out_dim):
    labels, curves = zip(*batch)
    labels = torch.stack(labels, dim=0)
    curves = torch.stack(curves, dim=0)

    pause = torch.zeros_like(curves[0])
    pause_block = torch.tile(pause, (num_pause_blocks, 1))
    curves_with_pause = torch.vstack([torch.vstack((row, pause_block)) for row in curves])
    visual_stimulus = TimedArray(curves_with_pause.numpy() * nA, dt=msec_step)

    stimulus_indexes = TimedArray(np.tile([1] + [0] * num_pause_blocks, curves.size(0)), dt=msec_step)

    out_stimuli_idx = get_output_stimuli_indexes_from_labels(labels, num_classes_per_layer)
    out_stimuli_array = get_output_current_arrays(out_stimuli_idx, out_dim, 1)
    label_pause = torch.zeros(out_dim)
    label_pause_block = torch.tile(label_pause, (num_pause_blocks, 1))
    
    out_stimuli_with_pause = torch.vstack([torch.vstack((row, label_pause_block)) for row in out_stimuli_array])
    output_stimuli = TimedArray(out_stimuli_with_pause.numpy() * nA, dt=msec_step)
    
    #out_stimuli_with_pause = np.zeros_like(torch.vstack([torch.vstack((row, label_pause_block)) for row in out_stimuli_array]))
    #output_stimuli = TimedArray(out_stimuli_with_pause * nA, dt=msec_step)


    return stimulus_indexes, output_stimuli, visual_stimulus

###############################################################################
# Build the network and run
###############################################################################
#dims = [22*22, 250, 125, 64]
N = 22
dims = [N*N,20,20,20]
num_classes_per_layer = 3
max_depth = 3

# Actually create the SNN
# Rs, Es_0, Es_1, G, Os, S_o_internal, connections = make_network(dims, num_classes_per_layer, max_depth)
Rs, Es_0, Es_1, G, Os, connections = make_network(dims, num_classes_per_layer, max_depth)

Rs[0].indices = range(Rs[0].N)

# Create the dataset
su = Synthetic_Dataset_Utils()
ranges = [30, 30, 30, 30]
prior_params = [10, 10, 10, 10]
num_samples_per_class = 500

tree = su.build_tree(prior_params, 0, max_depth, num_classes_per_layer, 1, ranges)
synth_dataset = su.make_dataset(tree, num_classes_per_layer, num_samples_per_class=num_samples_per_class, N=N)

output_dim = Os.N
# lr = 0.03
lr = 0.0002
batch_size = 40
minibatch_size = 5
num_pause_blocks = 1

dataset = CurveDataset(synth_dataset, 0.3, 1.2)
dl = DataLoader(
    dataset,
    shuffle=True,
    batch_size=batch_size,
    collate_fn=partial(
        collate_fn,
        msec_step=msec_step,
        num_pause_blocks=num_pause_blocks,
        num_classes_per_layer=num_classes_per_layer,
        out_dim=output_dim
    )
)



torch.save(dataset, "data/dataset1.pt")

#dataset = torch.load("data/dataset1.pt")




@network_operation(dt=msec_step, when='end')
def debug():
    pass
    #print("Rs[0].I: ", Rs[0].I[:])
    
    
def update_batch_sum(synapses):
    synapses.batch_sum_X += synapses.X  # Perform the operation
    
# In order to be able to update the weights in batches, we keep track of the cumulative value of X over a minibatch   
@network_operation(dt=msec_step, when='end')
def update_sums():
    current_idx = int(defaultclock.t / msec_step)
    if stimulus_indexes.values[current_idx] == 1:
        for n in range(len(Rs)):
            if n > 0:
                Rs[n].batch_sum_X += Rs[n].X
            Es_0[n].batch_sum_X += Es_0[n].X
            Es_1[n].batch_sum_X += Es_1[n].X
            
    Os.batch_sum_X += Os.X
            

# Used to regularize only non-zero, positive weights
def if_positive(x, w):
    # Return x where w > 0, otherwise 0 (element-wise)
    return np.where(w > 0, x, 0 * volt)



# This important function implements hebbian learning. It updates the weigght every msec_step*minibatch_size
debug_weights = []
@network_operation(dt=msec_step*minibatch_size, when='end')
def apply_weight_update():
    print("Applying weight update")
    for n in range(len(Rs)-1):
        
        print("n: ", n)
        
        # Retrieve top_down synapses       
        S_p, S_m = connections[f"top_down_{n}"]
        S_p_2, S_m_2 = connections[f"top_down_2_{n}"]
        
        # Multiply the (unitless) accumulated X of each synapse pre and post synaptic neurons, and use that to determine the update. If the weight
        # is positive, it tends to decrease (regularizer).
        for j_idx in range(Es_0[n].N):
            for i_idx in range(Rs[n+1].N):
                
                positive_connections_delta = lr * Es_0[n].batch_sum_X_[j_idx] * Rs[n+1].batch_sum_X_[i_idx] - if_positive(regularizer, S_p.w_[i_idx, j_idx])
                negative_connections_delta = lr * Es_1[n].batch_sum_X_[j_idx] * Rs[n+1].batch_sum_X_[i_idx] - if_positive(regularizer, S_m.w_[i_idx, j_idx])
                
                S_p.w[i_idx, j_idx] += positive_connections_delta/minibatch_size  * siemens
                S_m.w[i_idx, j_idx] -= negative_connections_delta/minibatch_size * siemens
                
                S_p_2.w[i_idx, j_idx] += positive_connections_delta/minibatch_size * siemens
                S_m_2.w[i_idx, j_idx] -= negative_connections_delta/minibatch_size * siemens
                
                
                    
            
            # Reset accumulated X after update
            Rs[n+1].batch_sum_X[:] = 0 * volt
            Es_0[n].batch_sum_X[:] = 0 * volt
            Es_1[n].batch_sum_X[:] = 0 * volt

        # Debug
        print("hehd: ", n+1)
        if (n+1) == 1:
            print("iuhichseiuhsi: ", positive_connections_delta)
            print("iuhichsesfei: ", negative_connections_delta)
            debug_weights.append(S_p.w)
    
    
    
    external_syn_p = connections["output_external_p"][0]
    external_syn_2_p = connections["output_external_2_p"][0]
    
    external_syn_m = connections["output_external_m"][0]
    external_syn_2_m = connections["output_external_2_m"][0]
    
    
    # Update synapses between last error layer and Output layer
    for i_idx in range(Es_0[-1].N):
        for j_idx in range(Os.N):
            
            delta_p = lr * Es_0[-1].batch_sum_X_[i_idx] * Os.batch_sum_X_[j_idx] * siemens
            external_syn_p.w[i_idx, j_idx] += delta_p/minibatch_size
            external_syn_2_p.w[i_idx, j_idx] += delta_p/minibatch_size
            
            delta_m = lr * Es_1[-1].batch_sum_X_[i_idx] * Os.batch_sum_X_[j_idx] * siemens
            external_syn_m.w[i_idx, j_idx] += delta_m/minibatch_size
            external_syn_2_m.w[i_idx, j_idx] += delta_m/minibatch_size
            
            
    Es_0[-1].batch_sum_X[:] = 0
    Es_1[-1].batch_sum_X[:] = 0
    Os.batch_sum_X[:] = 0
    
    # Automatic non-negativity for all Synapses objects
    #print("CHUNGUS: ")
    for obj in net.objects:
        if isinstance(obj, Synapses):
            # Zero out negative weights
            #print("Chungus")
            #print("A")
            #print(obj.w)
            #print("Negative weights:")
            #print(obj.w[obj.w < 0 * nsiemens])
            obj.w_[:] = np.clip(obj.w_[:], 0, None)
            #print("B")
            #print(obj.w)
        
            
# The following is to be sure to construct a functioning Network object            
     
# Extract all synapses and objects for the network
all_synapses = []
# all_synapses.append(S_o_internal)

for ii in range(len(Rs)-1):
    bu_syns = connections[f"bottom_up_{ii}"]
    all_synapses.extend(bu_syns)
    td_key = f"top_down_{ii}"
    if td_key in connections:
        td_syns = connections[td_key]
        all_synapses.extend(td_syns)

for ii in range(1,len(Rs)-1):
    bu_syns = connections[f"bottom_up_2_{ii}"]
    all_synapses.extend(bu_syns)
    td_key = f"top_down_2_{ii}"
    if td_key in connections:
        td_syns = connections[td_key]
        all_synapses.extend(td_syns)

gist_input_syn = connections["gist_input"][0]
gist_output_syn = connections["gist_output"][0]
all_synapses.append(gist_input_syn)
all_synapses.append(gist_output_syn)

external_syn_p = connections["output_external_p"][0]
external_syn_m = connections["output_external_m"][0]

external_syn_2_p = connections["output_external_2_p"][0]
external_syn_2_m = connections["output_external_2_m"][0]

all_synapses.append(external_syn_p)
all_synapses.append(external_syn_m)

all_synapses.append(external_syn_2_p)
all_synapses.append(external_syn_2_m)

monitor_Es1_a = StateMonitor(Es_1[0], ['v'], record=[0])
monitor_Es1_b = StateMonitor(Es_1[0], ['I_tot'], record=[0])
#spike_monitor = SpikeMonitor(Rs[0])

s = 0

        



monitor_o = StateMonitor(Os, ['v', 'I_tot', 'X'], record=True)
spike_mon_o = NamedSpikeMonitor(Os, record = True, custom_name = "output")

'''
net = Network(
    Rs, Es_0, Es_1, G, Os,
    all_synapses,        
    monitor_o, spike_mon_o,         # <-- Add these
    debug, update_sums, apply_weight_update
)
'''

# Add spike monitors for all neuron groups

spike_mon_G = NamedSpikeMonitor(G, record=True, custom_name = "Gist")

stimulus_indexes, output_stimuli, input_stimuli = next(iter(dl))

debug_monitor = NamedSpikeMonitor(Es_0[1], record=True, custom_name = f"Es_0_[1]_debug")
net = Network(
    Rs, Es_0, Es_1, G, Os,  # Neuron groups       
    monitor_o, spike_mon_o, debug_monitor,monitor_Es1_a,monitor_Es1_b, # Output monitors
    spike_mon_G,  # Spike monitors
    debug, update_sums, apply_weight_update # Additional operations
)

for l in range(len(Rs)):
    net.add(NamedSpikeMonitor(Rs[l], record=True, custom_name = f"R[{l}]"))
   
    net.add(NamedSpikeMonitor(Es_0[l], record=True, custom_name = f"Es_0[{l}]"))
    net.add(NamedSpikeMonitor(Es_1[l], record=True, custom_name = f"Es_1[{l}]"))

for synapse in all_synapses:
    net.add(synapse)

all_batch_rates = []
output_spikes = []

s = 0

#stimulus_indexes, output_stimuli, input_stimuli = next(iter(dl))

#net.stimulus_indexes = stimulus_indexes
#net.output_stimuli   = output_stimuli
#net.input_stimuli    = input_stimuli

run_time = (batch_size + batch_size*num_pause_blocks)*100*ms
net.run(run_time)
result_dir = f'results_v2'
#device.build(run=True, directory="project_v2",results_directory=result_dir) 

# device.build(directory='project', run = False)

states = net.get_states()
print("FUUUUUUUUUUUUCK")
b2t.brian_plot(monitor_Es1_a)
plt.show()
b2t.brian_plot(monitor_Es1_b)
plt.show()
keys = []
for key in states.keys():
    if "spike" in key:
        keys.append(key)
        print(key)
        
for key in keys:
    spike_mon_o = net[key]
    spikes = spike_mon_o.spike_trains()
    
    b2t.brian_plot(spike_mon_o)
    # Plotting the spike train with dots
    plt.figure(figsize=(10, 6))
    ii=0
    old = 0
    for neuron, spike_times in spikes.items():
    
        old = spike_times
        
        if ii < 100:
            plt.scatter(spike_times/ms, [neuron] * len(spike_times), s=10, label=f'Neuron {neuron}' if neuron == 0 else "")
            ii+=1

    plt.xlabel('Time (ms)')
    plt.ylabel('Neuron index')
    plt.title(f'{spike_mon_o.custom_name}. Num neurons = {len(spikes.items())}')
    print(spike_mon_o.source)
    plt.yticks(range(N))
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
    
```
